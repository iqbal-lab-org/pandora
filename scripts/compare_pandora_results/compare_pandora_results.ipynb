```python
import pandas as pd
```

```python
def load_pandora_matrix(filepath):
    pandora_matrix = pd.read_csv(filepath, sep="\t")
    pandora_matrix.rename(columns={'Unnamed: 0': 'locus'}, inplace=True)
    pandora_matrix.set_index("locus", inplace=True)
    return pandora_matrix
```

```python
pandora_matrix_lazy_loading = load_pandora_matrix("pandora_multisample.matrix.lazy_loading")
pandora_matrix_lazy_loading
```

```python
pandora_matrix_pre_lazy_loading = load_pandora_matrix("pandora_multisample.matrix.pre_lazy_loading")
pandora_matrix_pre_lazy_loading
```

```python
def get_sum_of_presence(df):
    return df.agg(['sum'])

sum_lazy_loading = get_sum_of_presence(pandora_matrix_lazy_loading)
sum_pre_lazy_loading = get_sum_of_presence(pandora_matrix_pre_lazy_loading)
all_sums = pd.concat([sum_lazy_loading, sum_pre_lazy_loading])
all_sums = all_sums.transpose()
all_sums.columns=["sum_lazy_loading", "sum_pre_lazy_loading"]
all_sums
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
ax = all_sums.plot.scatter(x='sum_lazy_loading', y='sum_pre_lazy_loading')
ax.plot([0, 1], [0, 1], transform=ax.transAxes)
ax.set_title("Number of genes present in each sample, pre and post lazy loading")
ax.set_xlim(0, 2000)
ax.set_ylim(0, 2000)
```

```python
def get_loci_present(df, sample):
    loci_present = set()
    for locus, presence in df[sample].to_dict().items():
        if presence:
            loci_present.add(locus)
    return loci_present
```

```python
from matplotlib_venn import venn3

for sample in pandora_matrix_lazy_loading.columns:
    venn3([get_loci_present(pandora_matrix_lazy_loading, sample),
           get_loci_present(pandora_matrix_pre_lazy_loading, sample)], ('Lazy_loading', 'Pre_lazy_loading'))

plt.show()
```
